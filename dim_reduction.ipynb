```python
import scipy as scp
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise_scale = 20
def gene_1(latent_state):
    return np.random.normal(np.exp(latent_state),scale = np.abs(np.exp(latent_state)/noise_scale))
def gene_2(latent_state):
    return np.random.normal(np.cos(latent_state),scale = np.abs(np.cos(latent_state)/noise_scale))
    return 
def gene_3(latent_state):
    return np.random.normal(np.sin(latent_state),scale = np.abs(np.sin(latent_state)/noise_scale))

def gene_4(latent_state):
    return np.random.normal(1 - latent_state**2,scale = np.abs((1 - latent_state**2)/noise_scale))
    return 
```

```python
latent_state = 6*np.random.rand(20)
```

```python
plt.scatter(gene_2(latent_state),gene_3(latent_state))
```

```python
def model(X, a1, a2, a3, a4, b):
    g1, g2, g3, g4 = X
    return g1*a1 + g2*a2 + g3*a3 + g4*a4 + b
```

```python
X = (gene_1(latent_state),gene_2(latent_state),gene_3(latent_state),gene_4(latent_state))
```

```python
a1, a2, a3, a4, b = scp.optimize.curve_fit(model,X, latent_state, p0=[0.2,1,1,1,1])[0]
```

```python
a1
```

```python
plt.scatter(model(X, a1, a2, a3, a4, b), latent_state)
```

```python
def model1(g1, a1, b):
    return g1*a1 + b
```

```python
a1, b = scp.optimize.curve_fit(model1,gene_1(latent_state), latent_state, p0=[0.2,1])[0]
```

```python
plt.scatter(model1(gene_1(latent_state), a1, b), latent_state)
```
