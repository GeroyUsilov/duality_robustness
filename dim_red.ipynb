```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps = 1000
its = 1000
dims = 100
results = []
for i in range(its):
    result = np.zeros((steps,dims))
    for j in range(1,steps):
        result[j,:] = result[j-1,:] + .1*np.random.normal(size = dims)
    results.append(np.array(result))
```

```python
results[0].shape
```

```python
from sklearn.decomposition import PCA
```

```python
# Perform PCA
first_d = []
second_d = []
for i in range(its):
    pca = PCA()  # Reduce to 2 principal components for visualization
    data_pca = pca.fit_transform(np.array(results[i]))
    first_d.append(pca.explained_variance_ratio_[0])
    second_d.append(pca.explained_variance_ratio_[1])


```

```python
plt.hist(first_d)
plt.show()
```

```python
plt.hist(np.array(first_d)+np.array(second_d),bins = 50)
plt.show()
```
