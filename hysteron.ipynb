```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
ns = [10000]
for j in range(len(ns)):
    n = ns[j]
    if True:
        h = np.random.rand(2,n)
        loc = 0.5
        scale = .3
        a, b = (0 - loc) / scale, (1 - loc) / scale
        h[0,:] = stats.truncnorm.rvs(a, b, loc=loc, scale=scale, size=n, random_state=None)
        h[1,:] = stats.truncnorm.rvs(a, b, loc=loc, scale=scale, size=n, random_state=None)
        h_p = np.zeros(n)
        h_m = np.zeros(n)
        h_p[h[0,:]>h[1,:]] = h[0,h[0,:]>h[1,:]]
        h_p[h[0,:]<h[1,:]] = h[1,h[0,:]<h[1,:]]
        h_m[h[0,:]>h[1,:]] = h[1,h[0,:]>h[1,:]]
        h_m[h[0,:]<h[1,:]] = h[0,h[0,:]<h[1,:]]
    if False:
        h_m = np.random.rand(n)
        h_p = np.array(h_m) + 0.5
        h_p = h_p/np.max(h_p)
        h_m = h_m/np.max(h_p)
      
    t = 1000
    m_up = np.zeros(t)
    m_down = np.zeros(t)
    steps_up = np.linspace(0,1,t)
    steps_down= np.linspace(1,0,t)

    for i in range(len(steps_up)):
        m_up[i] = np.mean(steps_up[i] > h_p)
    for i in range(len(steps_up)):
        m_down[i] = np.mean(steps_down[i] > h_m)
    plt.plot(steps_up,m_up,color = colors[j])
    plt.plot(steps_down,m_down,color = colors[j],ls = '--')
plt.ylabel('M')
plt.xlabel('H')
```
