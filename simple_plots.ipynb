```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(theta,l):
    return (1+l**2 - 2*l*np.sin(theta))/(np.cos(theta)**2)
```

```python
def find_min(l):
    thetas = np.linspace(-1,1,1000)
    results = f(thetas,l)
    return thetas[np.argwhere(results == np.min(results))][0]
```
