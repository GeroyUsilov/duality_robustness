```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
N = 50 # Number of particles
L = 10.0 # Box size
k = 1.0 # Spring constant
T = 1.0 # Temperature
dt = 0.01 # Time step
nsteps = 1000 # Number of time steps

# Initialize positions and velocities
positions = L*np.random.rand(N,2)
velocities = np.zeros((N,2))

# Initialize spring constants and equilibrium distances
K = k*np.random.rand(N,N)
K = 0.5*(K + K.T) # Make symmetric
np.fill_diagonal(K, 0) # Set diagonal to zero
D = np.linalg.norm(positions[:,np.newaxis,:] - positions, axis=2)

# Initialize plot
plt.ion()
fig, ax = plt.subplots()

# Loop over time steps
for step in range(nsteps):
    
    # Calculate forces
    dx = positions[:,np.newaxis,:] - positions
    r = np.linalg.norm(dx, axis=2)
    r = np.where(r == 0, 1, r) # Avoid divide-by-zero
    f = np.sum(-k*(r - D)*(dx.T/r),axis = 1).T
    np.fill_diagonal(f, 0) # Set diagonal to zero

    # Update velocities and positions using Verlet algorithm
    velocities += 0.5*dt*f
    positions += dt*velocities
    positions = np.mod(positions, L) # Apply periodic boundary conditions
    D = np.linalg.norm(positions[:,np.newaxis,:] - positions, axis=2)
    dx = positions[:,np.newaxis,:] - positions
    r = np.linalg.norm(dx, axis=2)
    r = np.where(r == 0, 1, r) # Avoid divide-by-zero
    f = np.sum(-k*(r - D)*(dx.T/r),axis = 1).T
    np.fill_diagonal(f, 0) # Set diagonal to zero

    velocities += 0.5*dt*f
    
    # Plot positions
    ax.clear()
    ax.scatter(positions[:,0], positions[:,1])
    ax.set_xlim(0, L)
    ax.set_ylim(0, L)
    plt.pause(0.001)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
# Set up the parameters
Lx = Ly = 1   # Length of domain in x and y directions
T = .1        # Total simulation time
nx = ny = 500  # Number of grid points in x and y directions
nt = 100      # Number of time steps
dx = Lx / nx
dy = Ly / ny
dt = T / nt
D = 0.1       # Diffusion coefficient

# Set up the initial conditions
u = np.zeros((nt+1, nx+1, ny+1))
u[0, :, :] = np.sin(np.pi*np.linspace(0, Lx, nx+1))[:, np.newaxis] \
                * np.sin(np.pi*np.linspace(0, Ly, ny+1))[np.newaxis, :]

# Define the finite difference scheme
for k in tqdm(range(nt)):
    for i in range(1, nx):
        for j in range(1, ny):
            u[k+1, i, j] = u[k, i, j] \
                + D*dt/dx**2 * (u[k, i+1, j] - 2*u[k, i, j] + u[k, i-1, j]) \
                + D*dt/dy**2 * (u[k, i, j+1] - 2*u[k, i, j] + u[k, i, j-1])
```
